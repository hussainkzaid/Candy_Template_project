# cleaning

### 4- drop all the useless columns and  rows with too much NaN ,
> tip : useless columns can be like `Internal ID`and can be also columns with too much NaN
- rename columns for easy access

```python
data = data.drop(['Unnamed: 113','Q12: MEDIA [Yahoo]','Q12: MEDIA [Daily Dish]','Q12: MEDIA [ESPN]','Q9: OTHER COMMENTS','Internal ID'],axis = 1)
data
```

```python
data = data[data.isnull().sum(axis = 1) <= 100].reset_index()
data.head()
```

### 5 - clean `going out` column
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
print(data['Q1: GOING OUT?'].value_counts(dropna=False))

mode = data['Q1: GOING OUT?'].mode()[0]

data['Q1: GOING OUT?'] = data['Q1: GOING OUT?'].fillna(mode)

print("**************************************")

print(data['Q1: GOING OUT?'].value_counts(dropna=False))
```

### 6 - clean ` gender` column
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
print(data['Q2: GENDER'].value_counts(dropna = False).sort_values(ascending=False))

mode = data['Q2: GENDER'].mode()[0]

data['Q2: GENDER'] = data['Q2: GENDER'].replace(np.nan, "I'd rather not say")

print(data['Q2: GENDER'].value_counts().sort_values(ascending=False))
```

### 7 - clean ` age` column
> tip : note the type of age column , and NaN values
 - try to use `to_numeric`
 - tip : you should choose which best value to put in place of these NaNs ( mean, median, ....etc)

```python

print(data['Q3: AGE'].value_counts(dropna = False))

data.loc[data['Q3: AGE'].isin(['I can remember when Java was a cool new language','MY NAME JEFF'])]

data = data.drop(data.loc[data['Q3: AGE'].isin(['I can remember when Java was a cool new language','MY NAME JEFF'])].index)

data.loc[data['Q3: AGE'].isin(['I can remember when Java was a cool new language','MY NAME JEFF'])]

data['Q3: AGE'] = data['Q3: AGE'].replace(['old','OLD','ancient','Over 50',"older than dirt"], '55')
data['Q3: AGE'] = data['Q3: AGE'].replace(['46 Halloweens.','old enough',"Old enough",'24-50','Enough',"Many","45-55"], '46')

data[data['Q3: AGE'] == 'See question 2']
data[data['Q3: AGE'].isin([100,1000])]
data["Q3: AGE"] = data["Q3: AGE"].replace(["See question 2",102,100,1000,"?","no","hahahahaha"], np.nan)

data["Q3: AGE"] = pd.to_numeric(data["Q3: AGE"])
mean_Age = data["Q3: AGE"].mean()

data["Q3: AGE"] = data["Q3: AGE"].replace(np.nan,mean_Age)
print("*****************************************")
print(data['Q3: AGE'].value_counts(dropna = False))
```
