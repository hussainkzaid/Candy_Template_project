### 7 - clean ` age` column
> tip : note the type of age column , and NaN values
 - try to use `to_numeric`
 - tip : you should choose which best value to put in place of these NaNs ( mean, median, ....etc)

```python
print(data['Q3: AGE'].value_counts(dropna = False))

data = data.drop(data.loc[data['Q3: AGE'].isin(['I can remember when Java was a cool new language','MY NAME JEFF'])].index)

data['Q3: AGE'] = data['Q3: AGE'].replace(['old','OLD','ancient','Over 50',"older than dirt"], '55')
data['Q3: AGE'] = data['Q3: AGE'].replace(['46 Halloweens.','old enough',"Old enough",'24-50','Enough',"Many","45-55"], '46')
data['Q3: AGE'] = data['Q3: AGE'].replace('60+', '60')
data['Q3: AGE'] = data['Q3: AGE'].replace('sixty-nine', '69')
data['Q3: AGE'] = data['Q3: AGE'].replace(['59 on the day after Halloween','5u'], '59')
data["Q3: AGE"] = data["Q3: AGE"].replace(["See question 2",102,100,1000,"?","no","hahahahaha",'your mom'], np.nan)

data["Q3: AGE"] = pd.to_numeric(data["Q3: AGE"])
data.loc[(data['Q3: AGE'] < 10) | (data['Q3: AGE'] > 100) , 'Q3: AGE'] = np.nan
mean_Age = data["Q3: AGE"].mean()
data["Q3: AGE"] = data["Q3: AGE"].replace(np.nan, mean_Age)
print("*****************************************")
data["Q3: AGE"] = data["Q3: AGE"].round().astype(int)
data['Q3: AGE'].value_counts(dropna = False)
```

### 8 - clean ` country` column
> tip : note that country column has country names with different writing ways
  - use fuzzywuzzy to deal with this problem
  - you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
!pip install fuzzywuzzy
!pip install pycountry
import fuzzywuzzy
from fuzzywuzzy import fuzz, process
import pycountry

data["Q4: COUNTRY"] = data["Q4: COUNTRY"].apply(lambda x: "Unknown" if isinstance(x, int) else x)

data["Q4: COUNTRY"] = data["Q4: COUNTRY"].fillna("Unknown")
data["Q4: COUNTRY"].unique()
countries = [c.name for c in pycountry.countries]
abr_dic = {
    'i pretend to be from canada, but i am really from the united states': "United States",
    "murica": "United States",
    'the united states of america': "United States",
    "unknown": "Unknown",
    "uae": "United Arab Emirates",
    "us": "United States","usa": "United States","ussa": "United States","u.s.": "United States","u.s.a.": "United States",
    "u s": "United States","u s a": "United States","usausausa": "United States",
    "america": "United States",
    'murrica': "United States",'murrika': "United States","'merica": "United States",
    'us of a':"United States",
    "ahemamerca": "United States",
    "usa? hard to tell anymore": "United States",'trumpistan': "United States",
    "usa usa usa!!!!":"United States",
    'pittsburgh':"United States",'new york':"United States",'new jersey':"United States",'north carolina':"United States",
    'california':"United States",'alaska':"United States",
    "uk": "United Kingdom",
    "u.k.": "United Kingdom",
    "britain": "United Kingdom",
}

def clean_country(name, choices=countries, threshold=60):
    raw = str(name).strip()
    key = (raw.lower().replace(".", "")).strip()

    if key in abr_dic:
        return abr_dic[key]

    match, score = process.extractOne(raw, choices, scorer=fuzz.token_sort_ratio)

    if score >= threshold:
        return match
    else:
        return "Unknown"

data['Q4: COUNTRY'] = data['Q4: COUNTRY'].apply(clean_country)
data['Q4: COUNTRY'].value_counts(dropna = False)
```

### 14 - write a summary about all the cleaning steps

### Summary of Cleaning Steps

**Starting point:** 2,460 rows and 120 columns, with a lot of messiness to sort through.

**1. Media columns**
Combined the four separate media columns into one `Media` column, picking whichever one had a value. Rows with nothing in any of them got labeled "Unknown."

**2. Dropping unnecessary data**
- Removed columns that weren't useful (`Unnamed: 113`, the comments column, the despair-other column) since they were mostly empty
- Dropped rows with too many missing values to be worth keeping
- Converted `Internal ID` to a string, since it's an identifier, not something to do math on

**3. Going Out & Gender**
Filled missing values with the mode for Going Out, and "I'd rather not say" for Gender.

**4. Age**
The messiest column by far. People typed things like "old enough," "60+," "MY NAME JEFF," and "your mom." Steps taken:
- Manually mapped decipherable answers to actual numbers
- Removed entries that were clearly jokes
- Converted the column to numeric
- Filtered out unrealistic ages (under 10 or over 100)
- Filled remaining missing values with the average age

**5. Country**
Same info written a dozen different ways (US, USA, us, America, Murica, etc.). Handled it by:
- Mapping the obvious variants with a dictionary
- Using fuzzy matching against real country names for everything else
- Defaulting to "Unknown" when nothing matched well

**6. State/Province**
Same idea as country — grouped abbreviations and typos into their proper names manually.

**7. Candy ratings (Q6 columns)**
Instead of cleaning each candy column individually, reshaped that whole section into a long-format table (one row per person per candy). Dropped blank ratings, since a missing rating just means the person skipped that candy — not something that needs filling in.

**8. Dress & Day**
Barely needed touching. Day had a few "Unknown" values, replaced with the mode.

**Result:** Main dataset now has correct data types, no leftover garbage values, and no missing data that matters — with candy ratings split into their own table, ready to be joined back in for analysis.
